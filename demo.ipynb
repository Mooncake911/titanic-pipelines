```python
explore_column = "Transported"
index_column = 'PassengerId'

EXPLORE_MODEL = os.environ.get('EXPLORE_MODEL')
print(EXPLORE_MODEL)
```

```python
df = pd.read_csv(filepath_or_buffer="data/result.csv", low_memory=False)

split_index = df[explore_column].isnull().idxmax() 
train = norm_data(df=df.loc[:split_index-1], ignore_columns=[index_column])
test = norm_data(df=df.loc[split_index:], ignore_columns=[index_column, explore_column])

display(train.tail(10))
display(test.head(10))
```

# Drop columns

```python
train = train.sample(frac=1, ignore_index=True)  # shuffle 
train = train.drop(columns=['PassengerId', 'Cabin', 'Name'])
# Correlation
sns.heatmap(train.corr(), annot=True, cmap='RdYlGn', linewidths=0.1, vmin=0)
# Prediction
models, _, _ = check_models(train, pred_column=explore_column)
```

# Prediction

```python
name_models = models.keys()
print(name_models)

if EXPLORE_MODEL not in name_models:
    raise ValueError(f"Model not found in {name_models}!")
```

```python
x_train = train.drop(explore_column, axis=1)
y_train = train[explore_column]
x_test = test[x_train.columns]
# y_test = test[pred_column]

model = models[EXPLORE_MODEL]
```

```python
data = pd.DataFrame()
data[index_column] = test[index_column].values
data[explore_column] = model.predict(x_test)
data[explore_column] = data[explore_column].replace({1: True, 0: False})
display(data)

data.to_csv('data/output.csv', index=False)
```
