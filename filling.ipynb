```python
import pandas as pd
```

```python
explore_column = "Transported"
```

```python
train = pd.read_csv(filepath_or_buffer="data/train.csv", low_memory=False)
train.info()
train.head(10)
```

```python
if explore_column not in train.columns:
    raise ValueError(f"{explore_column} not in data!")
```

```python
test = pd.read_csv(filepath_or_buffer="data/test.csv", low_memory=False)
test.info()
test.head(10)
```

```python
result = pd.merge(train, test, on=test.columns.tolist(), how='outer')
result.info()
result.head(10)
```

```python
def fill_most_frequent(df):
    for column in df.columns:
        if column != explore_column:
            most_common_value = df[column].mode().iloc[0]
            df[column] = df[column].fillna(most_common_value)
    return df
```

```python
result = fill_most_frequent(result)    
result.info()
result.to_csv('data/result.csv', index=False)
```
